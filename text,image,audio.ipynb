```python
mp3_file=r'/content/WhatsApp Audio 2025-05-21 at 10.36.21 AM.mp3'
from pydub import AudioSegment
from pydub.playback import play
audio=AudioSegment.from_file(mp3_file)
audio.export("temp2.wav",format="wav")
```

```python
import speech_recognition as sr
import re

# Step 1: Load and convert audio to text
model = sr.Recognizer()
audio_file = r'/content/temp2.wav'

with sr.AudioFile(audio_file) as source:
    audio_data = model.record(source)
    text = model.recognize_google(audio_data)
    print("Extracted Text:", text)

# Step 2: Use regex to detect and evaluate a simple addition expression
pattern = r'(\d+)\s*\+\s*(\d+)'
match = re.search(pattern, text)

if match:
    num1 = int(match.group(1))
    num2 = int(match.group(2))
    result = num1 + num2
    print(f"Result: {result}")
else:
    print("No valid addition expression found.")
```
